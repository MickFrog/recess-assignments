```python
#Equality 
equality = series1 == series2
print("Equality comparison")
print(equality)
```

```python
#Greater than
greater_than = series1 > series2
print("Greater than comparison")
print(greater_than)
```

```python
# SERIES NO 2
# redefine the same series declared before
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

# Addition
addition_result = series1 + series2
print("Addition Result:")
print(addition_result)

# Subtraction
subtraction_result = series1 - series2
print("Subtraction Result:")
print(subtraction_result)

# Multiplication
multiplication_result = series1 * series2
print("Multiplication Result:")
print(multiplication_result)

# Division
division_result = series1 / series2
print("Division Result:")
print(division_result)
```

```python
# SERIES NO 5
afcon_df = pd.read_csv("AfricaCupofNationsMatches.csv")

homeTeamSeries = afcon_df["HomeTeamGoals"]

#get most frequent value in series
mst_frequent_homeTeam = homeTeamSeries.value_counts().idxmax()
print(mst_frequent_homeTeam)
```
